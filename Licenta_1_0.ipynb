```python
import pandas as pd
import string
import re
data_set_1 = pd.read_csv('../hashtag_joebiden.csv')
data_set_1 = data_set_1.iloc[:,[0]] # Pastrez doar coloana cu tweet-uri
data_set_1.to_csv(index = False)
data_set_1.head(10)
```

```python
# @title Eliminam simbolurile si semnele de punctuatie
def eliminate_punctuations(text):
  punctuations = string.punctuation
  return text.translate(str.maketrans('','',punctuations))
```

```python
data_set_1 = data_set_1.apply(lambda x: eliminate_punctuations(x))
data_set_1.head()
```
